## Stock Sentiment Analysis using News Headlines ( Lemmatization, Stopwords Applied)

```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
df=pd.read_csv('Data.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
# Training and testing data
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

### Training Data

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head()
```

```python
' '.join(str(x) for x in data.iloc[2,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer # Steming 
from nltk.stem import WordNetLemmatizer # Lemmatization
```

```python
lematizer = WordNetLemmatizer()
corpus = []
for i in range(len(headlines)):
    words = nltk.word_tokenize(headlines[i])
    words = [lematizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentence = ' '.join(words)
    corpus.append(sentence)
```

```python
corpus[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

### Bag of Words

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(corpus)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

### For Testing Data 

```python
# Removing punctuations
data=test.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index

# Converting into lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head()
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(data.index)):
    test_transform.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
test_transform[0]
```

```python
lematizer = WordNetLemmatizer()
corpus_test = []
for i in range(len(test_transform)):
    words = nltk.word_tokenize(test_transform[i])
    words = [lematizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentence = ' '.join(words)
    corpus_test.append(sentence)
```

```python
corpus_test[0]
```

```python
test_dataset = countvector.transform(corpus_test)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
